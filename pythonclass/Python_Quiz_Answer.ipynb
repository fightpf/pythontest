```python
# Q2
def generator_fib(max_num):
    a, b = 0, 1
    while a < max_num:
        yield a
        a, b = b, a + b
gen_fib = generator_fib(max_num=10)
checker.ex2_1(generator_fib=gen_fib)
```

```python
# Q3
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w *self.h
class Student:
    def calculate(self, rec1, rec2):
        return rec1.area() + rec2.area()
    
rect1 = Rectangle(2,3)
rect2 = Rectangle(3,4)
student = Student()

checker.ex2_2(student, rect1, rect2)
```

```python
# Q4
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
iris = datasets.load_iris()
iris_data = pd.DataFrame(data=iris['data'],columns=iris['feature_names'])
iris_data["target"] = iris['target']
```

```python
column_names = iris_data.columns
plt.figure(figsize=(15,15))
for row in range(5):
    for col in range(5):
        plt.subplot(5,5, row*5+col+1)
        if row == col:
            plt.hist(iris_data.iloc[:,col])
        else:
            color_wheel = {1: "#FF0000", 
            2: "#00FF00", 
            3: "#0000FF"}
            colors = iris_data["target"].map(lambda x: color_wheel.get(x + 1))
            plt.scatter(iris_data.iloc[:,col], iris_data.iloc[:,row], c=colors, alpha=0.5, marker='.')
        if col == 0:
            plt.ylabel(column_names[row])
        if row == 4:
            plt.xlabel(column_names[col])
            
# use scatter_matrix
from pandas.tools.plotting import scatter_matrix
ax = scatter_matrix(iris_data, color=colors, alpha=0.6, figsize=(15, 15), diagonal='hist')
```
